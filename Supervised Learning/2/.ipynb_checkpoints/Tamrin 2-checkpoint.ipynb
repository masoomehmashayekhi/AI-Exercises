```python
import numpy as np
import pandas as pd
import math
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score

raw_data = pd.read_csv("./BOM.csv", skiprows=1, names=range(23))

col_name = raw_data.columns[22]
raw_data = raw_data.dropna(subset=[col_name])

raw_data.pop(0)
disc_attr = [1, 7, 9, 10, 21, 22]
cont_attr = []

for i in range(1, 22):
    if i not in disc_attr:
        cont_attr.append(i)
#print(raw_data.isna().sum())

for col in range(1, 22):
    if col in cont_attr:
        mean_val = raw_data[col].mean()
        raw_data.loc[:, col] = raw_data[col].replace(math.nan, mean_val)
    if col in disc_attr:
        mod_val = raw_data[col].mode()[0]
        raw_data.loc[:, col] = raw_data[col].replace(math.nan, mod_val)

#print(raw_data.isna().sum())
y = raw_data.pop(22)

lbl_encoder = LabelEncoder()
y = lbl_encoder.fit_transform(y)

disc_attr.pop(5)

hot_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
new_data = hot_encoder.fit_transform(raw_data[disc_attr])
x = pd.DataFrame(new_data)
x.index = raw_data.index
raw_data = raw_data.drop(disc_attr, axis=1)

X = pd.concat([raw_data, x], axis=1)

min_max_scaler = MinMaxScaler()
X = min_max_scaler.fit_transform(X)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

k_change = range(1, 20)
founded_n = 0
score = 0
for n in k_change:
    knn = KNeighborsClassifier(n_neighbors=n)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    score1 = f1_score(y_test, y_pred)
    if score1 > score:
        founded_n = n
        score = score1
    print("score1 : " + str(score1))
    print("score : " + str(score))
    print("n : " + str(n))
    print("founded_n : " + str(founded_n))

#بهترین امتیاز برای k=5 
#درخت تصمیم برای حالت real_time بهتر است
#زیرا الگوریتم نزدیکترین همسایه الکوریتم تنبل است و تمام محاسبات در زمان پیش بینی انجام می شود
#اما درخت تصمیم برای یادگیری براساس داده کارهایی انجام میدهد و تمام بار محاسبات به زمان پیش بینی اختصاص نمی یابد
```
