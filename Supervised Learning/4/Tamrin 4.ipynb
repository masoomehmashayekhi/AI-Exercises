```python
import ssl
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score 

ssl._create_default_https_context = ssl._create_unverified_context

faces = fetch_olivetti_faces(shuffle=True, random_state=42)

X = faces.data        # (400, 64*64)
y = faces.target      # (400,)
print("X shape:", X.shape)
print("y shape:", y.shape)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2)

plt.figure(figsize=(12, 6))
for i in range(40):
    plt.subplot(4, 10, i+1)
    img = X[y == i][0].reshape(64, 64)
    plt.imshow(img, cmap='gray')
    plt.axis('off')
    plt.title(f"Person {i}", fontsize=8)
plt.suptitle("one pic of each person", fontsize=16)
plt.show()

svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)
acc_svm = accuracy_score(y_test, y_pred_svm)
print("SVM Accuracy:", acc_svm)

rf_model = RandomForestClassifier(n_estimators=100)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
acc_rf = accuracy_score(y_test, y_pred_rf)
print("Random Forest Accuracy:", acc_rf)


ada_model = AdaBoostClassifier(n_estimators=100)
ada_model.fit(X_train, y_train)
y_pred_ada = ada_model.predict(X_test)
acc_ada = accuracy_score(y_test, y_pred_ada)
print("AdaBoost Accuracy:", acc_ada)

#  پیشنهادات برای بهبود نتایج
#
#1. استفاده از مدل‌های پیچیده‌تر مثل CNN برای استخراج ویژگی‌ها.
#2. افزایش داده‌ها با Data Augmentation (چرخش، تغییر نور، کراپ و ...).
#3. پیش‌پردازش بهتر تصاویر (نرمال‌سازی و حذف نویز).
#4. استفاده از ترکیب چند مدل (Ensemble) یا Fine-tuning مدل‌های از قبل آموزش‌دیده
```
