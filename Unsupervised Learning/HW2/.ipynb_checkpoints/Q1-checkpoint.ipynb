```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
filename = 'diabetes.csv'
df = pd.read_csv(filename)
```

```python
Y = df.pop('Outcome')
X = df
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
print(np.isnan(X_scaled).any()) 
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
lda = LDA(n_components=1)   
X_lda = lda.fit_transform(X_scaled, Y)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=Y, cmap='viridis', s=10)
plt.title('PCA projection of diabetes dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

plt.figure(figsize=(8,5))
plt.scatter(X_lda, [0]*len(X_lda), c=Y, cmap='viridis', s=10)
plt.title('LDA projection of diabetes dataset')
plt.xlabel('Linear Discriminant 1')
plt.yticks([])
plt.show()
```

```python
def pca_numpy(X, n_components=2):
     
    X_centered = X - np.mean(X, axis=0)
 
    cov_matrix = np.cov(X_centered, rowvar=False)
 
    eigvals, eigvecs = np.linalg.eigh(cov_matrix)
 
    sorted_indices = np.argsort(eigvals)[::-1]
    eigvals = eigvals[sorted_indices]
    eigvecs = eigvecs[:, sorted_indices]
 
    W = eigvecs[:, :n_components]
 
    X_pca = np.dot(X_centered, W)

    return X_pca, eigvals, W
```

```python
def lda_numpy(X, y, n_components=1): 
    X = np.asarray(X)
    y = np.asarray(y)
    class_labels = np.unique(y)
    n_features = X.shape[1]
 
    mean_overall = np.mean(X, axis=0)

    S_W = np.zeros((n_features, n_features))
    S_B = np.zeros((n_features, n_features))

    for c in class_labels:
        X_c = X[y == c]
        mean_c = np.mean(X_c, axis=0)
         
        S_W += np.dot((X_c - mean_c).T, (X_c - mean_c))
         
        n_c = X_c.shape[0]
        mean_diff = (mean_c - mean_overall).reshape(n_features, 1)
        S_B += n_c * np.dot(mean_diff, mean_diff.T)

     
    eigvals, eigvecs = np.linalg.eig(np.linalg.pinv(S_W).dot(S_B))

    
    sorted_indices = np.argsort(abs(eigvals))[::-1]
    eigvecs = eigvecs[:, sorted_indices]
    eigvals = eigvals[sorted_indices]

     
    W = eigvecs[:, :n_components]

     
    X_lda = np.dot(X, W.real)
    return X_lda
```

```python
pca_nmp = pca_numpy(X,n_components=2)
lda_nmp = lda_numpy(X, Y, n_components=1)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=Y, cmap='viridis', s=10)
plt.title('PCA sklearn projection of diabetes dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

plt.figure(figsize=(8,6))
plt.scatter(pca_nmp[0][:,0], pca_nmp[0][:,1], c=Y, cmap='viridis', s=10)
plt.title('PCA numpy projection of diabetes dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

plt.figure(figsize=(8,5))
plt.scatter(X_lda, [0]*len(X_lda), c=Y, cmap='viridis', s=10)
plt.title('LDA sklearn projection of diabetes dataset')
plt.xlabel('Linear Discriminant 1')
plt.yticks([])
plt.show()


plt.figure(figsize=(8,5))
plt.scatter(lda_nmp, [0]*len(lda_nmp), c=Y, cmap='viridis', s=10)
plt.title('LDA numpy projection of diabetes dataset')
plt.xlabel('Linear Discriminant 1')
plt.yticks([])
plt.show()
```
