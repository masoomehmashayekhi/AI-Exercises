```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(0, 100, size=(1, 50))  
y = np.random.randint(0, 100, size=(1, 50)) 
```

```python
distances = []
for dim in range(2, 51):
    d = np.linalg.norm(x[:, :dim] - y[:, :dim]) 
    distances.append(d)
print(distances)
```

```python
plt.scatter(range(2, 51),distances)
plt.plot(range(2, 51),distances, color='red')

plt.title("Euclidean distance vs Number of Dimensions")
plt.xlabel("x = Number of Dimensions")
plt.ylabel("y = Euclidean distance")
```
